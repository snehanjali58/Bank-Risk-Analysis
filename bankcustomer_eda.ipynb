# Exploratory Data Analysis 

```python
# Creating a income bin column from existing a column income.

bins=[0,100000,300000,float('inf')]
labels=['Low','Med','High']

df['Income Band']= pd.cut(df['Estimated_Income'], bins=bins,labels=labels, right=False)
print(df['Income Band'])
```

```python
#Plotting a Bar chart of count of Income Band using matplotlib

df1= df.groupby(['Income Band'], as_index=False)['Client_ID'].count()
df1
```

```python
plt.bar(df1['Income Band'], df1['Client_ID'], color='Cyan', edgecolor='Black')
plt.xlabel('Income Band')
plt.title('Count of Income Band')
plt.show()
```

```python
#Plotting the same thing using Countplot of seaborn
ax=sns.countplot(data=df, x='Income Band', hue='Nationality')

for bars in ax.containers:
    ax.bar_label(bars, fmt='{:.0f}')
```

```python
#Exploring the distribution of the Categorical columns of the dataset.

cat_col=["Nationality","Fee_Structure","Loyalty_Classification","Amount_of_Credit_Cards","Properties_Owned","Risk_Weighting","BRId","GenderId","IAId"]

for col in cat_col:
    print(f"Value counts of the {col}")
    print(df[col].value_counts(),"\n")
```

```python
#Plotting this distribution of each categorical column using countplot of the seaborn. 
#This purely comes under univariate analysis.

for index,col in enumerate(cat_col):
    plt.figure(index)
    sns.countplot(data=df, x=col)
```

```python
#Bivariate analysis. Under this we are trying to find the relation or pattern betwenn the 2 columns. 

for index, col in enumerate(cat_col):
    plt.figure(index)
    ax=sns.countplot(data=df, x=col,hue='Nationality' )
    for container in ax.containers:
        ax.bar_label(container, fmt='%d', label_type='edge')
```

```python
# Creating subplots in single figure

# Setup subplots
n_cols = 2
n_rows = (len(cat_col) + n_cols - 1) // n_cols
fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 5 * n_rows))
axes = axes.flatten()

# Plot each categorical distribution
for i, col in enumerate(cat_col):
    sns.countplot(x=col, data=df, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')
    #axes[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels for readability
```

```python
# Numerical Analysis:- We will perform some univariate analysis onto numerical column.
# To see the distribution of numerical analysis. We have to use histograms. 

num_col= ["Estimated_Income","Superannuation_Savings" ]

fig,axes= plt.subplots(2,1, figsize=(8,10))
axes=axes.flatten()

for i,col in enumerate(num_col):
    sns.histplot(data=df, x=col, kde=True,ax=axes[i])
    
```

```python
# Now we will take a look at the some of heatmaps. To find out the correlation between all numerical columns.

num_col2= ["Estimated_Income","Superannuation_Savings","Credit_Card_Balance","Bank_Loans","Bank_Deposits","Checking_Accounts","Saving_Accounts","Foreign_Currency_Account","Business_Lending"]

corr_matrix=df[num_col2].corr()

plt.figure(figsize=(8,8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
